```python
from ultralytics import YOLO
import os

pathhh = 'brain-tumor/' 
model = YOLO("yolov8x.pt")

model.train(
    data=os.path.join(pathhh, 'data.yaml'), 
    epochs=15,
    imgsz=640,
    batch=4,
    name='brain_tumor_yolov8x',
    patience=20,
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    degrees=10.0,
    translate=0.1,
    scale=0.5,
    shear=2.0,
    flipud=0.2,
    fliplr=0.5,
    mosaic=1.0,
    mixup=0.2
)
```

```python
import os

# Path to the weights directory
weights_dir = r'C:\Users\engqu\dp\dp\runs\detect\brain_tumor_yolov8x\weights'
os.makedirs(weights_dir, exist_ok=True)

# Create dummy .pt files if they don't exist
for file_name in ['best.pt']:
    file_path = os.path.join(weights_dir, file_name)
    if not os.path.exists(file_path):
        with open(file_path, 'wb') as f:
            f.write(b'')  # Create empty .pt file
        print(f"Created dummy file: {file_path}")
    else:
        print(f"File already exists: {file_path}")
```

```python
model.save(os.path.join(weights_dir, 'best.pt'))
```

```python
from ultralytics import YOLO

# Load the best model
model = YOLO('/best.pt')

# Test prediction (replace with your image path)
results = model.predict(r"C:\Users\engqu\dp\dp\datasets\brain-tumor\valid\images\val_1 (6).jpg")
results[0].show()
```

```python
metrics = model.val()
# Access mean precision and mean recall using the methods mp() and mr()
print("Precision:", metrics.box.mp) # Corrected from metrics.box.precision
print("Recall:", metrics.box.mr) # Corrected from metrics.box.recall
# Access mean AP@0.5 using the map50() method
print("mAP@0.5:", metrics.box.map50) # Corrected from metrics.box.map50
# Access mean AP@0.5:0.95 using the map() method
print("mAP@0.5:0.95:", metrics.box.map) # Corrected from metrics.box.map
```
